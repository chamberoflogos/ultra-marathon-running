```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Clean df3 deterministically ---
df3 = df3.copy()

# 1) numeric y
df3["athlete_average_speed"] = pd.to_numeric(df3["athlete_average_speed"], errors="coerce")

# 2) keep only the two race lengths shown in the tutorial
df3["race_length"] = df3["race_length"].astype(str).str.strip()
df3 = df3[df3["race_length"].isin(["50km", "50mi"])]

# 3) normalize gender to M/F only (so hue has exactly two levels)
df3["athlete_gender"] = df3["athlete_gender"].astype(str).str.strip().str[0].str.upper()
df3 = df3[df3["athlete_gender"].isin(["M", "F"])]

# 4) drop impossible/garbage speeds (keeps the same visual range as the tutorial)
df3 = df3[df3["athlete_average_speed"].between(2.5, 18.5)]

# --- Plot: four side-by-side violins with quartiles, fixed y-range ---
ax = sns.violinplot(
    data=df3,
    x="race_length",
    y="athlete_average_speed",
    hue="athlete_gender",
    split=False,          # side-by-side (gives 4 violins)
    dodge=True,
    inner="quartile",     # <- correct keyword
    cut=0,
    scale="width",
    order=["50km", "50mi"],
    hue_order=["M", "F"],
)
ax.set_ylim(2.5, 18.5)
ax.set_xlabel("race_length")
ax.set_ylabel("athlete_average_speed")
plt.show()


```

```python
#Are runners slower in the summer compared to the winter?
#Spring 3-5 
#Summer 6-8
#Fall 9-11
#Winter 12-2
```

```python
df3 = df3.copy()
df3['race_month'] = (
    df3['race_day'].astype(str).str.split('.').str[1].astype(int)
)

season_map = {12:'Winter', 1:'Winter', 2:'Winter',
              3:'Spring', 4:'Spring', 5:'Spring',
              6:'Summer', 7:'Summer', 8:'Summer',
              9:'Fall', 10:'Fall', 11:'Fall'}

df3['race_season'] = df3['race_month'].map(season_map)

```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'] \
   .agg(['mean', 'count']) \
   .sort_values('mean', ascending=False)
```
